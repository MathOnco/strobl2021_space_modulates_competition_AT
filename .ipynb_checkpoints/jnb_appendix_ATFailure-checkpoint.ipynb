## Bar Plots with % of Failed Runs

### No Cost, No Turnover

```python
# Load the data
txComparisonDf_noTurnover = pd.read_csv('./data/n0Sweep_noTurnover.csv',index_col=0)
txComparisonDf_noTurnover = txComparisonDf_noTurnover[txComparisonDf_noTurnover.Turnover==False]
txComparisonDf_noTurnover['InitialTumourSize'] *= 100
txComparisonDf_noTurnover.InitialTumourSize = txComparisonDf_noTurnover.InitialTumourSize.astype(int)
```

```python
txComparisonDf_noTurnover['AbsTimeGained_CA'] = txComparisonDf_noTurnover.TTP_AT50_CA-txComparisonDf_noTurnover.TTP_CT_CA
txComparisonDf_noTurnover['SignificantLoss_30d'] = (txComparisonDf_noTurnover.AbsTimeGained_CA<-30)*100
txComparisonDf_noTurnover['SignificantLoss_10p'] = (txComparisonDf_noTurnover.RelTimeGained_CA<-10)*100
txComparisonDf_noTurnover['Cured'] = ((txComparisonDf_noTurnover.FinalSize_R_CT==0)&(txComparisonDf_noTurnover.FinalSize_R_AT>0))*100
txComparisonDf_noTurnover['SignificantLoss'] = ((txComparisonDf_noTurnover.SignificantLoss_30d==100)|
                                                  (txComparisonDf_noTurnover.SignificantLoss_10p==100)|
                                                  (txComparisonDf_noTurnover.Cured==100))*100
```

```python
fig, axList = plt.subplots(1,1,figsize=(10,8))
currAx = axList
sns.barplot(x='InitialTumourSize',y='SignificantLoss',#ci='sd',
            hue="RFrac", palette=sns.xkcd_palette(['coral','scarlet','dark red']),
            edgecolor='black',linewidth=2,
            data=txComparisonDf_noTurnover,ax=currAx)
currAx.set_ylim(0,10)
currAx.set_xlabel("")
currAx.set_ylabel("")
currAx.tick_params(labelsize=32)
currAx.get_legend().remove()
sns.despine(offset=5, trim=True)
plt.xticks(rotation=45);
plt.tight_layout()
plt.savefig("figures/failures_noTurnover_noCost.pdf")
```

### With Cost, With Turnover

```python
# Load the data
txComparisonDf_withCost = pd.read_csv('./data/n0Sweep_withCost.csv',index_col=0)
txComparisonDf_withCost = txComparisonDf_withCost[txComparisonDf_withCost.Cost==0.3]
txComparisonDf_withCost['InitialTumourSize'] *= 100
txComparisonDf_withCost.InitialTumourSize = txComparisonDf_withCost.InitialTumourSize.astype(int)
print(txComparisonDf_withCost.shape)
# Remove those runs in which no progression occured in 10,000 days
txComparisonDf_withCost = txComparisonDf_withCost[(np.isnan(txComparisonDf_withCost.TTP_CT_CA)==False)&(np.isnan(txComparisonDf_withCost.TTP_AT50_CA)==False)]
print(txComparisonDf_withCost.shape)
```
