```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import sys
import os
from tqdm import tqdm
import imageio
import cv2
import scipy
from itertools import product

sys.path.append("../../utils")
from myUtils import mkdir
from odeAnalysisUtils import Simulate_AT_FixedThreshold,rdModel_nsKill,GenerateParameterDic
from caAnalysisUtils import RunSimulation, GenerateATComparisonPlot_CA, GenerateATComparisonPlot_ODE

# Format plots
sns.set(style="white",
        font_scale=2.,
        font="Serif",
        rc={'figure.figsize':(12,6)})
```

## Example Simulations

```python
# Params for the figure
initialSize = 0.5
rFrac = 0.01
turnoverList = [0,0.3]
cost = 0
nReplicates = 250
tEnd = 475
dataDir = "./data/exampleSims_odeComparison/"
```

### Collect Data

#### ODE

```python
if runCASimulations:
    for turnover, in tqdm(product(turnoverList)):
        currOutDir = os.path.join(dataDir,"ODE/turnover%1.2f/cost%1.2f/"%(turnover,cost))
        mkdir(currOutDir)
        # Get the parameters
        initialStateVec,currParamDic = GenerateParameterDic(initialSize=initialSize,rFrac=rFrac,
                                                            cost=cost,turnover=turnover,paramDic=paramDic)

        for txName in ["MTD", "AT50"]:
            # Simulate
            resultsDf = Simulate_AT_FixedThreshold(initialStateVec,modelFun=rdModel_nsKill,
                                                   paramDic=currParamDic,
                                                   atThreshold=1. if txName=="MTD" else 0.5,
                                                   intervalLength=intervalLength,
                                                   t_end=tEnd,
                                                   t_eval=np.arange(0,tEnd+dt,dt))
            currDfName = os.path.join(currOutDir,"%s_cellCounts_rFrac_%.2g_initSize_%.2g_dt_%1.1f_RepId_0.csv" % (txName, rFrac, initialSize, dt))
            resultsDf.to_csv(currDfName)
```

#### CA

```python
# Collect the data for the time series plots showing the treatment dynamics avaraged across nReplicate simulations
if runCASimulations:
    for turnover, in tqdm(product(turnoverList)):
        currOutDir = os.path.join(dataDir,"CA/turnover%1.2f/cost%1.2f/"%(turnover,cost))
        modelConfigDic = {"initialSize":initialSize,"rFrac":rFrac,"turnover":turnover,"cost":cost,
                          "tEnd":tEnd,"nReplicates":nReplicates,
                          "profilingMode":"false","terminateAtProgression":"false",
                          "outDir":currOutDir}
        RunSimulation(modelConfigDic,jarFileName='./onLatticeModel.jar')
```

### Plot ODE Plots

```python
for turnover in turnoverList:
    currOutDir = os.path.join(dataDir,"ODE/turnover%1.2f/cost%1.2f/"%(turnover,cost))
    # Read the data
    tmpList = []
    replicateId = 0
    for txName in ["MTD","AT50"]:
        currDfName = os.path.join(currOutDir,"%s_cellCounts_rFrac_%.2g_initSize_%.2g_dt_%1.1f_RepId_0.csv" % (txName, rFrac, initialSize, dt))
        tmpDf = pd.read_csv(currDfName,index_col=0)
        tmpDf['TreatmentBranch'] = txName
        tmpDf['ReplicateId'] = replicateId
        tmpList.append(tmpDf)
    dataDf = pd.concat(tmpList)
    dataDf.rename(columns={"D":"DrugConcentration"},inplace=True)
    # Plot
    GenerateATComparisonPlot_CA(dataDf,t_end=tEnd,ylim=1.3,savefigB=True,
                                decoratex=False,decoratey=turnover==turnoverList[0],
                                outName="figures/odeVsCAComparison_ODE_turnover%1.2f.pdf"%(turnover),
                                figsize=(8,5))
```

### Plot CA Plots

```python
for turnover in turnoverList:
    currOutDir = os.path.join(dataDir,"CA/turnover%1.2f/cost%1.2f/"%(turnover,cost))
    # Read the data
    tmpList = []
    for replicateId in range(nReplicates): #nReplicates
        for txName in ["MTD","AT50"]:
            currDfName = os.path.join(currOutDir,"%s_cellCounts_cost_%1.1f_rFrac_%.2g_initSize_%.2g_dt_1.0_RepId_%d.csv"%(txName,cost*100,rFrac,initialSize,replicateId))
            tmpDf = pd.read_csv(currDfName)
            tmpDf['NCells'] /= 1e4
            tmpDf['NCells_S'] /= 1e4
            tmpDf['NCells_R'] /= 1e4
            tmpDf['TreatmentBranch'] = txName
            tmpDf['ReplicateId'] = replicateId
            tmpList.append(tmpDf)
    dataDf = pd.concat(tmpList)
    dataDf.rename(columns={"NCells":"V","NCells_S":"S","NCells_R":"R"},inplace=True)
    # Plot
    GenerateATComparisonPlot_CA(dataDf,minTime=150,t_end=tEnd,ylim=1.3,savefigB=True,
                                decoratex=True,decoratey=turnover==turnoverList[0],
                                outName="figures/odeVsCAComparison_CA_turnover%1.2f.pdf"%(turnover),
                                figsize=(8,5))
```

## Comparison of TTP between ODE and ABM

### Generate ttp data from ODE

```python
initialSizeList = [0.35, 0.5, 0.65]
rFracList = [0.1, .01, .001]
costList = [0, 0.1]
turnoverList = [0, 0.3]
tEnd = 10*365
outDir = "./data/odeData/"
```

#### Simulate ODE

```python
for initialSize, rFrac, cost, turnover in tqdm(list(product(initialSizeList,rFracList,costList,turnoverList))):
    currOutDir = os.path.join(outDir,"turnover%d"%(turnover*100))
    currOutDir = os.path.join(currOutDir,"cost%d"%(cost*100))
    mkdir(currOutDir)
    # Get the parameters
    initialStateVec,currParamDic = GenerateParameterDic(initialSize=initialSize,rFrac=rFrac,
                                                        cost=cost,turnover=turnover,paramDic=paramDic)

    for txName in ["MTD", "AT50"]:
        # Simulate
        resultsDf = Simulate_AT_FixedThreshold(initialStateVec,modelFun=rdModel_nsKill,
                                               paramDic=currParamDic,
                                               atThreshold=1. if txName=="MTD" else 0.5,
                                               intervalLength=intervalLength,
                                               t_end=tEnd,
                                               t_eval=np.arange(0,tEnd+dt,dt))
        currDfName = os.path.join(currOutDir,"%s_cellCounts_rFrac_%.2g_initSize_%.2g_dt_%1.1f_RepId_0.csv" % (txName, rFrac, initialSize, dt))
        resultsDf.to_csv(currDfName)
```

#### Assemble data frame

```python
# Load large scale parameter sweep data (n=1000). Generated using the 'collectData_odeComparison.py' script
txComparisonDf = pd.read_csv("data/n0Sweep_odeComparison.csv",index_col=0)
txComparisonDf['TTP_MTD_ODE'] = np.nan
txComparisonDf['TTP_AT50_ODE'] = np.nan
txComparisonDf.drop(columns=['FinalSize_R_AT','FinalSize_R_CT','SuppressionDiff_Sq','SuppressionDiff_Sum','SuppressionDiff_Abs'],inplace=True)
```

```python
for initialSize, rFrac, cost, turnover in tqdm(list(product(initialSizeList,rFracList,costList,turnoverList))):
    currOutDir = os.path.join(outDir,"turnover%d"%(turnover*100))
    currOutDir = os.path.join(currOutDir,"cost%d"%(cost*100))
    for txName in ["MTD", "AT50"]:
        currDfName = os.path.join(currOutDir,"%s_cellCounts_rFrac_%.2g_initSize_%.2g_dt_%1.1f_RepId_0.csv" % (txName, rFrac, initialSize, dt))
        tmpDf = pd.read_csv(currDfName,index_col=0)
        ttp = tmpDf.Time[tmpDf.V > 1.2*tmpDf.V.iloc[0]].min()
        txComparisonDf.loc[(txComparisonDf.InitialTumourSize==initialSize) & (txComparisonDf.RFrac==rFrac) & 
                           (txComparisonDf.Cost==cost) & (txComparisonDf.Turnover==turnover),'TTP_%s_ODE'%txName] = ttp

# Compute the relative time gained
txComparisonDf['RelTimeGained_ODE'] = (txComparisonDf.TTP_AT50_ODE-txComparisonDf.TTP_MTD_ODE)/txComparisonDf.TTP_MTD_ODE*100
```

```python
txComparisonDf.to_csv("./data/txComparisonDf.csv")
```

### Plotting Code

```python
def QQPlot(dataDf,x="TTP_AT50_ODE",y="TTP_AT50_CA",axLimVec=None,palette=rFracPalette,sizeDic=sizeDic,ax=None):
    if ax==None: fig,ax = plt.subplots(1,1,figsize=(7,7))
    # Plot the lines
    sns.lineplot(x=x,y=y,ci='sd',err_style='band',
                 hue="RFrac", style="RFrac",
                 markers=True,markersize=1,
                 palette=palette,
                 ax=ax,data=dataDf)
    # Add the markers in different sizes to reflect n0
    summaryDf = dataDf.groupby(["RFrac","InitialTumourSize"]).mean()
    summaryDf.reset_index(inplace=True)
    sns.scatterplot(x=x,y=y, style='RFrac',
                    hue='RFrac', palette=palette,
                    size="InitialTumourSize",
                    markers=True, sizes=sizeDic,
                    ax=ax,
                    legend=False, data=summaryDf)
    # Set x and y axis to equal scale
    if axLimVec is None:
        xLimVec = ax.get_xlim()
        yLimVec = ax.get_ylim()
        axLimVec = [min(xLimVec[0],yLimVec[0]),max(xLimVec[1],yLimVec[1])]
    ax.set_xlim(axLimVec)
    ax.set_ylim(axLimVec)

    # Plot diagonal line to show where TTP would be equal
    xVec = np.linspace(*axLimVec,100)*0.9
    ax.plot(xVec,xVec,color=sns.xkcd_rgb['blue grey'],lw=3,ls='--')

    # Remove axis labels
    ax.set_xlabel("")
    ax.set_ylabel("")
    ax.tick_params(labelsize=24)
    ax.get_legend().remove()
    sns.despine(ax=ax,offset=50, trim=True)
```

## Understanding the difference in the absolute time to progression

```python
# Params for the figure
initialSize = 0.25
rFracList = [1., 0.1, .01, .001]
turnover = 0
cost = 0
nReplicates = 250
tEnd = 550
seedToShowImagesFor = 0
dataDir = "./data/exampleSims_singlePopulation/data/"
imgDir = "./data/exampleSims_singlePopulation/images/"
```

### Collect data

```python
# Collect the data for the time series plots showing the treatment dynamics avaraged across nReplicate simulations
if runCASimulations:
    for rFrac, in tqdm(product(rFracList)):
        currOutDir = os.path.join(dataDir,"turnover%1.2f/cost%1.2f/"%(turnover,cost))
        modelConfigDic = {"initialSize":rFrac*initialSize,"rFrac":1.,"turnover":turnover,"cost":cost,
                          "tEnd":tEnd,"nReplicates":nReplicates,"atThreshold":0.,
                          "compareToMTD":False,
                          "profilingMode":"false","terminateAtProgression":"false",
                          "outDir":currOutDir}
        RunSimulation(modelConfigDic,jarFileName='./onLatticeModel.jar')
```

```python
# Images
if runCASimulations:
    for rFrac, in tqdm(product(rFracList)):
        modelConfigDic = {"initialSize":rFrac*initialSize,"rFrac":1.,"turnover":turnover,"cost":cost,
                          "tEnd":tEnd,"seed":seedToShowImagesFor,"nReplicates":1,"atThreshold":0.,
                          "compareToMTD":False,
                          "profilingMode":"false","terminateAtProgression":"false",
                          "imageOutDir":imgDir,"imageFreq":10,
                          "outDir":"./data/scratch/"}
        RunSimulation(modelConfigDic,jarFileName='./onLatticeModel.jar')
```

### Assemble data frame

```python
fileNameDic = {1.:'0.25',0.1:'0.025',0.01:'0.0025',0.001:'2.5E-4'}
tmpList = []
for rFrac in rFracList:
    currOutDir = os.path.join(dataDir,"turnover%1.2f/cost%1.2f/"%(turnover,cost))
    # Read the data
    for replicateId in range(nReplicates): #nReplicates
        for txName in ["AT0"]:
            currDfName = os.path.join(currOutDir,"%s_cellCounts_cost_%1.1f_rFrac_1.0_initSize_%s_dt_1.0_RepId_%d.csv"%(txName,cost*100,fileNameDic[rFrac],replicateId))
            tmpDf = pd.read_csv(currDfName)
            tmpDf = tmpDf[['Time','NCells_R']]
            tmpDf['NCells_R'] /= 1e4
            tmpDf['RFrac'] = rFrac
            tmpDf['TreatmentBranch'] = txName
            tmpDf['ReplicateId'] = replicateId
            tmpList.append(tmpDf)
```

```python
def LogisticGrowth(t,r=0.1,K=1,n0=0.1):
    return K/(1+(1/(n0/K)-1)*np.exp(-r*t))

tmpDf['NCells_R'] = LogisticGrowth(tmpDf.Time,r=0.027,K=1e4,n0=1)/1e4
tmpDf['RFrac'] = -1
tmpList.append(tmpDf)
dataDf = pd.concat(tmpList)
```

```python
def LogTime(n,r=0.1,K=1,n0=0.1):
    return -1/r*np.log((K-n)/(K-n0)*n0/n)

dataDf['Time_Relative'] = dataDf.Time
for rFrac in rFracList: 
    # Change time to logistic reference time
    refTime = LogTime(n=int(rFrac*initialSize*1e4),r=0.027,K=1e4,n0=1)
    dataDf.loc[dataDf.RFrac==rFrac,'Time_Relative'] += refTime
dataDf = dataDf[dataDf.Time_Relative<dataDf.Time.max()]
```

### Plot

```python
ax = None
if ax==None: fig,ax = plt.subplots(1,1,figsize=(10,3)) #(10,7)
# Plot the lines
sns.lineplot(x="Time_Relative",y="NCells_R",ci='sd',
             hue="RFrac", style="RFrac",
             palette={**rFracPalette,1:"purple",-1:"black"},
             lw=5,
             ax=ax,data=dataDf)

# Remove axis labels
ax.set_xlabel("")
ax.set_ylabel("")
ax.tick_params(labelsize=16)
ax.get_legend().remove()
sns.despine(ax=ax,offset=5, trim=True)
plt.tight_layout()
plt.savefig("./figures/growthDynamicsComparison.pdf")
```
