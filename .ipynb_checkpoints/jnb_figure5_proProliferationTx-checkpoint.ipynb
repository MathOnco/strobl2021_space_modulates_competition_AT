```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
%matplotlib inline
import sys
import os
from tqdm import tqdm
import imageio
import cv2
import scipy
from itertools import product

sys.path.append("../../utils")
from myUtils import mkdir
from caAnalysisUtils import RunSimulation, GenerateATComparisonPlot_CA, GenerateATComparisonPlot_ODE, PlotCompetitionOverTime

# Format plots
sns.set(style="white",
        font_scale=2.,
        font="Serif",
        rc={'figure.figsize':(12,6)})
```

```python
### Script parameters
paramDic = {"rS":.027, "rR":.027, "cRS":1., "cSR":1.,"dD":1.5, 
            "k":1., "D":0, "theta":1, 'DMax':1.}
dt = 1.
tEnd = 1.1e3
nTimePts = 100
intervalLength = 1
colorDic = {'MTD':'#FF9409','AT50':'#094486'}
rFracPalette = {0.001:sns.xkcd_rgb['coral'],0.01:sns.xkcd_rgb['scarlet'],0.1:sns.xkcd_rgb['dark red']}
sizeDic = {0.35:150,0.5:350,0.65:550}
runCASimulations = True # Whether to run the CA simulations to generate the data analysed in this script
```

## Example Simulations

```python
# Params for the figure
initialSize = 0.5
rFrac = 0.01
turnover = 0.3
cost = 0
nReplicates = 250
tEnd = 750
proProliferationFacList = [0,0.3]
balanceList = [0,0.5]
dataDir = "./data/exampleSims_proProliferationTx/"
```

### Collect Data

```python
if runCASimulations:
    for proProliferationFac,balance in tqdm(product(proProliferationFacList,balanceList)):
        if proProliferationFac==0 and balance>0: continue # proProlifFac=0 is the no-proProlifTx control. Only need to simulate that for one balance value as there's
        currOutDir = os.path.join(dataDir,"proProlifFac%d/balance%1.2f/data/"%(proProliferationFac*100,balance))
        modelConfigDic = {"initialSize":initialSize,"rFrac":rFrac,"turnover":turnover,"cost":cost,
                          "tEnd":tEnd,"nReplicates":nReplicates,
                          "proProlifFac":proProliferationFac,"proProlifBalance":balance,
                          "profilingMode":"false","terminateAtProgression":"false",
                          "outDir":currOutDir}
        RunSimulation(modelConfigDic,jarFileName='./onLatticeModel.jar')
```

### Plot

```python
casesToShowList = [(0,0)]+[(x,y) for x in proProliferationFacList if x!=0 for y in balanceList]

for caseId,(proProliferationFac,balance) in enumerate(casesToShowList):
    currOutDir = os.path.join(dataDir,"proProlifFac%d/balance%1.2f/data/"%(proProliferationFac*100,balance))
    # Read the data
    tmpList = []
    for replicateId in range(nReplicates): #nReplicates
        for txName in ["MTD","AT50"]:
            currDfName = os.path.join(currOutDir,"%s_cellCounts_cost_%1.1f_rFrac_%.2g_initSize_%.2g_dt_1.0_RepId_%d.csv"%(txName,cost*100,rFrac,initialSize,replicateId))
            tmpDf = pd.read_csv(currDfName)
            tmpDf['NCells'] /= 1e4
            tmpDf['NCells_S'] /= 1e4
            tmpDf['NCells_R'] /= 1e4
            tmpDf['TreatmentBranch'] = txName
            tmpDf['ReplicateId'] = replicateId
            tmpList.append(tmpDf)
    dataDf = pd.concat(tmpList)
    dataDf.rename(columns={"NCells":"V","NCells_S":"S","NCells_R":"R"},inplace=True)
    atName = "AT50"
    if proProliferationFac>0:
        atName = "AT+PPTx"
        dataDf.TreatmentBranch.replace(to_replace="AT50",value="AT+PPTx",inplace=True)
    # Plot
    GenerateATComparisonPlot_CA(dataDf,minTime=150,t_end=tEnd,ylim=1.3,savefigB=True,
                                decoratex=True,decoratey=True,atName=atName,
                                outName="figures/proProlifTx_exampleSim_%d.pdf"%(caseId),
                                figsize=(8,5))
```

## Parameter Sweep

```python
proProlifScreeningDf = pd.read_csv("./data/proProliferationSweep_costTurnover.csv",index_col=0)
proProlifScreeningDf['AbsTimeGained'] = proProlifScreeningDf.TTP_AT50_CA-proProlifScreeningDf.TTP_CT_CA
proProlifScreeningDf['ProProlifFac'] *= 100
proProlifScreeningDf['ProProlifBalance'] *= 100
```

```python
initialSize = 0.5
rFrac = 0.01
cost = 0
turnover = 0.3

fig,ax = plt.subplots(1,1,sharex=True,sharey=True,figsize=(6,7))
currDataDf = proProlifScreeningDf[(proProlifScreeningDf.InitialTumourSize==initialSize) &
                                (proProlifScreeningDf.RFrac==rFrac) &
                                (proProlifScreeningDf.Turnover==turnover) &
                                (proProlifScreeningDf.Cost==cost)]

# Compute the mean TTP gained
summaryStatsDf = currDataDf[["ProProlifFac","ProProlifBalance","AbsTimeGained"]].groupby(["ProProlifFac","ProProlifBalance"]).mean()
summaryStatsDf = summaryStatsDf.reset_index(level=["ProProlifFac","ProProlifBalance"])
summaryStatsDf.rename(columns={"AbsTimeGained":"MeanAbsTimeGained"},inplace=True)
meanTimeGainedMat = summaryStatsDf.pivot("ProProlifBalance","ProProlifFac",'MeanAbsTimeGained')

# Plot
PlotTTPHeatmap(meanTimeGainedMat,cmap='viridis',vmin=0,vmax=85,cbar=False,ax=ax)
ax.invert_yaxis()
ax.set_yticklabels(['%d'%x for x in proProlifScreeningDf.ProProlifBalance.unique()])
ax.set_xticklabels(['%d'%x for x in proProlifScreeningDf.ProProlifFac.unique()])
plt.tight_layout()
plt.savefig("./figures/proProlifTx_paramSweep.pdf")
```

## Analyse the competition

```python
for caseId,(proProliferationFac,balance) in enumerate([(0,0),(0.3,0)]):
    currOutDir = os.path.join(dataDir,"proProlifFac%d/balance%1.2f/data/"%(proProliferationFac*100,balance))
    fig,ax = plt.subplots(1,1,figsize=(7,6))
    tmpList = []
    for i,txName in enumerate(["MTD","AT50"]):
        for replicateId in range(nReplicates):
            currDfName = os.path.join(currOutDir,"%s_cellCounts_cost_%1.1f_rFrac_%.2g_initSize_%.2g_dt_1.0_RepId_%d.csv"%(txName,cost*100,rFrac,initialSize,replicateId))
            tmpDf = pd.read_csv(currDfName)
            tmpDf['TreatmentBranch'] = txName
            tmpDf['ReplicateId'] = replicateId
            tmpList.append(tmpDf)
    dataDf = pd.concat(tmpList)
    dataDf.rename(columns={"NCells":"V","NCells_S":"S","NCells_R":"R"},inplace=True)
    # Plot
    dataDf = dataDf[dataDf.Time<1000]
    dataDf = dataDf[::2]
    dataDf['PropFailedDivs'] = dataDf.NFailedDivs/dataDf.NAttemptedDivs*100
    diffVec = dataDf.PropFailedDivs[dataDf.TreatmentBranch=="AT50"]-dataDf.PropFailedDivs[dataDf.TreatmentBranch=="MTD"]
    diffVec = diffVec[np.isnan(diffVec)==False]
    print(proProliferationFac,np.mean(diffVec))
    if proProliferationFac>0: dataDf.TreatmentBranch.replace(to_replace="AT50",value="AT+PPTx",inplace=True)
    PlotCompetitionOverTime(dataDf,decoratey=i>0,ylim=130,y2lim=1/1.2,ax=ax)
    plt.tight_layout()
    plt.savefig("figures/proProlifTx_competitionAnalysis_%d.pdf"%caseId)
```

### Quantify the peak inhibition

```python
tmpDicList = []
for caseId,(proProliferationFac,balance) in enumerate([(0,0),(0.3,0)]):
    currOutDir = os.path.join(dataDir,"proProlifFac%d/balance%1.2f/data/"%(proProliferationFac*100,balance))
    # Read the data
    tmpList = []
    txName = "AT50"
    for replicateId in range(nReplicates): #nReplicates
        currDfName = os.path.join(currOutDir,"%s_cellCounts_cost_%1.1f_rFrac_%.2g_initSize_%.2g_dt_1.0_RepId_%d.csv"%(txName,cost*100,rFrac,initialSize,replicateId))
        tmpDf = pd.read_csv(currDfName)
        tmpDf['PropFailedDivs'] = tmpDf.NFailedDivs/tmpDf.NAttemptedDivs*100
        tmpDf = tmpDf[::10]
        compResultsDf = tmpDf.groupby('NCycles').max()
        tmpDicList += [{"TreatmentBranch":proProliferationFac,"ReplicateId":replicateId,"CycleId":x[0],
                        "MaxBlockedDivs":x[1].PropFailedDivs} for x in compResultsDf.iterrows()]   
competitionComparisonDf = pd.DataFrame(tmpDicList)   
```

```python
fig, axList = plt.subplots(1,1,figsize=(8,10))
currAx = axList
sns.lineplot(x='CycleId',y='MaxBlockedDivs',err_style='bars',
             style='TreatmentBranch',hue='TreatmentBranch',
                 palette=sns.xkcd_palette(['blue','dark green']),
             markers=True, lw=3, markersize=25,
             data=competitionComparisonDf,ax=currAx)
#     currAx.set_ylim(-50,1.3e3)
currAx.set_xlabel("")
currAx.set_ylabel("")
currAx.tick_params(labelsize=32)
currAx.get_legend().remove()
sns.despine(offset=5, trim=True)
plt.xticks(rotation=45);
plt.savefig("figures/proProlifTx_maxBlockingComparison.pdf")
```
