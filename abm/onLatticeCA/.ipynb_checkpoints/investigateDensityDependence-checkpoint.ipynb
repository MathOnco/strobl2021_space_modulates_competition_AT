# Investigate the Density Dependence of Growth

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import imageio
from IPython.display import Image
import os
import sys
import pandas as pd
from tqdm import tqdm
import statsmodels.formula.api as sm
from scipy import stats
sys.path.append("../../../../assessingODEModels/odeFramework/")
import myUtils as utils
from referenceModelWrapper import ReferenceModelClass

# Format plot
sns.set(style="white",
        font_scale=1.5,
        rc={'figure.figsize':(12,6)})
```

## 1. Collect Data

### 1.1. Low Density

```python
paramDic={"rS":0.03,"rR":0.,
          "mS":0.0,"mR":0.0,
          "dS":0.,"dR":0.,
          "kS":0.} # Turn off movement but otherwise use cecile's parameters

lowDensityModel = ReferenceModelClass(paramDic=paramDic,dt=0.01,outFName="data_densityDependence/lowDensity.csv",
                               seed=42,loggingFreq=1.,verboseLevel=1, printFreq=100, visualise=True,
                                     imagingFreq=10,imageOutDir="data_densityDependence/images_lowDensity/")
lowDensityModel.SetInitialConditions(confluence=0.05,sensitiveFraction=1.)
lowDensityModel.Run([[0,220,0]],
             pause=.01)
lowDensityModel.PlotSimulation()
```

```python
images = []
fNameList = [float(x.split("_")[-1].split(".png")[0]) for x in os.listdir("data_densityDependence/images_lowDensity/")]
fNameList.sort()
for x in tqdm(fNameList):
    images.append(imageio.imread(os.path.join("data_densityDependence/images_lowDensity/","img_t_%1.1f.png"%x)))
imageio.mimsave('figures_densityDependence/lowDensitySimulation.gif', images, duration=.15)
```

```python
with open('figures_densityDependence/lowDensitySimulation.gif','rb') as f:
    display(Image(data=f.read(), format='png'))
```

### 1.2. High Density

```python
paramDic={"rS":0.03,"rR":0.,
          "mS":0.0,"mR":0.0,
          "dS":0.,"dR":0.,
          "kS":0.} # Turn off movement but otherwise use cecile's parameters

highDensityModel = ReferenceModelClass(paramDic=paramDic,dt=0.01,outFName="data_densityDependence/highDensity.csv",
                                       seed=42,loggingFreq=1.,verboseLevel=1, printFreq=100, visualise=True,
                                       imagingFreq=10,imageOutDir="data_densityDependence/images_highDensity/")
highDensityModel.SetInitialConditions(confluence=0.5,sensitiveFraction=1.)
highDensityModel.Run([[0,220,0]],
             pause=.1)
highDensityModel.PlotSimulation()
```

```python
images = []
fNameList = [float(x.split("_")[-1].split(".png")[0]) for x in os.listdir("data_densityDependence/images_highDensity/")]
fNameList.sort()
for x in tqdm(fNameList):
    images.append(imageio.imread(os.path.join("data_densityDependence/images_highDensity/","img_t_%1.1f.png"%x)))
imageio.mimsave('figures_densityDependence/highDensitySimulation.gif', images, duration=.15)
```

```python
with open('figures_densityDependence/highDensitySimulation.gif','rb') as f:
    display(Image(data=f.read(), format='png'))
```

## Plot Data

```python
dataDf_ld = pd.read_csv("data_densityDependence/lowDensity.csv")
dataDf_hd = pd.read_csv("data_densityDependence/highDensity.csv")

dataDf_ld['InitialDensity'] = 'low'
dataDf_hd['InitialDensity'] = 'high'
halfWayTime = dataDf_ld.Time[dataDf_ld.NCells>5000].min()
dataDf_hd['Time'] += halfWayTime
dataDf = pd.concat([dataDf_ld,dataDf_hd])
```

```python
fix,ax = plt.subplots(1,1,figsize=(12,8))
sns.lineplot(x="Time",y="NCells",hue="InitialDensity",style="InitialDensity",
             lw=7,legend=False,palette=sns.xkcd_palette(['sage','sea blue']),
             data=dataDf,ax=ax,
)

# palette=sns.xkcd_palette([] if seedingDensity=='lowDensity' else ['cornflower']),#,'pine green','mustard']),
ax.set_xlim(0,220)
```
